```python
model_1 = EncoderDecoderGRU.load_from_checkpoint(pretrained_weights_1).cpu()
model_2 = EncoderDecoderGRU.load_from_checkpoint(pretrained_weights_2).cpu()
```

```python
profile = np.random.choice(list(curve_dir.keys()))
sampling_rate = 0.5
title = f"Profile {profile} : Checkpoints comparison"
actual_current, actual_voltage, predicted_voltage_1 = profile_predict(
    model_1, curve_dir[profile]
)
_, _, predicted_voltage_2 = profile_predict(model_2, curve_dir[profile])

# Create time axis in minutes
time_axis = np.arange(len(actual_current)) / (sampling_rate * 60)

fig, ax = plt.subplots(2, figsize=(20, 8))
fig.suptitle(title)

# Plot actual and predicted voltages
ax[0].plot(time_axis, actual_current, label="Actual Current")
ax[1].plot(time_axis, actual_voltage, label="Actual Voltage")
ax[1].plot(time_axis, predicted_voltage_1, label="Predicted Voltage 1")
ax[1].plot(time_axis, predicted_voltage_2, label="Predicted Voltage 2")

# Calculate MAE
mae_1 = mean_absolute_error(actual_voltage, predicted_voltage_1)
mae_2 = mean_absolute_error(actual_voltage, predicted_voltage_2)

# Add R^2 score and MAE to legend
ax[1].legend(
    [
        "Actual Voltage",
        f"Predicted Voltage (MAE_1 : {mae_1:.3f}, MAE_2 : {mae_2:.3f})",
    ]
)

# Set axis label
ax[1].set_xlabel("Time (minutes)")
ax[0].set_ylabel("Current")
ax[1].set_ylabel("Voltage")
# ax[1].legend()

# Show plot
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(10,6))
# Compute MAE for each pair of encoder_voltage and predicted_voltage
mae_values = []
for encoder_voltage_element, predicted_voltage_element in zip(
    encoder_voltage_values, predicted_voltage
):
    mae = mean_absolute_error(
        np.full_like(predicted_voltage_element, encoder_voltage_element),
        predicted_voltage_element,
    )
    mae_values.append(mae)

# Plot MAE values against encoder_voltage_values
# plt.figure(figsize=(10, 6))
ax.plot(encoder_voltage_values, mae_values, marker="o", linestyle="-")
ax.set_ylim(0, 0.5)
ax.fill_betweenx(
    [0, 0.5], 4.12, 3.37, color="lightgray", alpha=0.5
)  # Shade the region from 4.12 to 3.37
ax.text(
    3.745, np.mean(mae_values), "Encoder Training Range", fontsize=12, ha="center"
)  # Add text annotation for the shaded area
ax.set_xlabel("Encoder Voltage")
ax.set_ylabel("Mean Absolute Error (MAE)")
ax.set_title("Relaxation Error(MAE) across the voltage range")
ax.grid(True)
ax.invert_xaxis()

num_predictions = 20
encoder_current_values = [
    0
] * num_predictions  # Replace with your list of encoder current values
encoder_voltage_values = np.linspace(
    4.15, 3, num_predictions
)  # Adjust voltage levels accordingly
decoder_current_values = [
    0
] * num_predictions  # Replace with your list of decoder current values

encoder_current, decoder_current, encoder_voltage, predicted_voltage = (
    perform_multiple_predictions(
        model_2,
        encoder_current_values,
        encoder_voltage_values,
        decoder_current_values,
        decoder_input_length=300,
    )
)

mae_values = []
for encoder_voltage_element, predicted_voltage_element in zip(
    encoder_voltage_values, predicted_voltage
):
    mae = mean_absolute_error(
        np.full_like(predicted_voltage_element, encoder_voltage_element),
        predicted_voltage_element,
    )
    mae_values.append(mae)

# Plot MAE values against encoder_voltage_values
# plt.figure(figsize=(10, 6))
ax.plot(encoder_voltage_values, mae_values, marker="o", linestyle="-")
ax.set_ylim(0, 0.5)
ax.fill_betweenx(
    [0, 0.5], 4.12, 3.37, color="lightgray", alpha=0.5
)  # Shade the region from 4.12 to 3.37
ax.text(
    3.745, np.mean(mae_values), "Encoder Training Range", fontsize=12, ha="center"
)  # Add text annotation for the shaded area
ax.set_xlabel("Encoder Voltage")
ax.set_ylabel("Mean Absolute Error (MAE)")
ax.set_title("Relaxation Error(MAE) across the voltage range")
ax.grid(True)
ax.invert_xaxis()
```

```python
"""
Compute dV/dI by keeping the encoder voltage fixed and varying the current
Goal : It should be constant. 
Method : Compute the std of the whole set of values
"""
fig, ax = plt.subplots(1, figsize=(10, 5))

_decoder_voltage = np.array([element[0] for element in predicted_voltage])

iR_drop = encoder_voltage_values - _decoder_voltage
ax.scatter(
    decoder_current_values,
    iR_drop,
    label=f"Encoder Voltage = {encoder_voltage_values[0]:.2f}",
)
ax.fill_betweenx([-0.05, 0.2], 2.5, 0.75, color="lightgray", alpha=0.5)
fig.suptitle("Pure Ohmic Drop")
ax.set_xlabel("Current")
ax.set_ylabel("Pure Ohmic Drop")
ax.legend()
ax.grid("True")

num_predictions = 20
encoder_current_values = [0] * num_predictions
encoder_voltage_values = [3.8] * num_predictions
decoder_current_values = np.linspace(0.25, 3.5, num_predictions)
encoder_current, decoder_current, encoder_voltage, predicted_voltage = (
    perform_multiple_predictions(
        model_2,
        encoder_current_values,
        encoder_voltage_values,
        decoder_current_values,
        decoder_input_length=300,
    )
)

plot_prediction(
    encoder_current,
    decoder_current,
    encoder_voltage,
    predicted_voltage,
    title=f"Discharge Behavior at {num_predictions} different SOC/voltage levels",
)
_decoder_voltage = np.array([element[0] for element in predicted_voltage])

iR_drop = encoder_voltage_values - _decoder_voltage
ax.scatter(
    decoder_current_values,
    iR_drop,
    label=f"Encoder Voltage = {encoder_voltage_values[0]:.2f}",
)
ax.fill_betweenx([-0.05, 0.2], 2.5, 0.75, color="lightgray", alpha=0.5)
fig.suptitle("Pure Ohmic Drop")
ax.set_xlabel("Current")
ax.set_ylabel("Pure Ohmic Drop")
ax.legend()
ax.grid("True")
```

```python
"""
Compute the slope of the discharge curve by keeping the encoder voltage fixed and varying the current
Expectation : It should increase (abs) as the current increases.
Reality : 
"""
fig, ax = plt.subplots(1, figsize=(10, 5))

start = int(predicted_voltage[0].shape[0] / 3)

slopes = []
for i in range(num_predictions):
    slopes.append(np.polyfit(np.arange(start), predicted_voltage[i][-start:], 1)[0])
ax.scatter(
    decoder_current_values,
    slopes,
    label=f"Encoder Voltage = {encoder_voltage_values[0]:.2f}",
)


ax.fill_betweenx([-0.0012, 0], 2.5, 0.75, color="lightgray", alpha=0.5)
fig.suptitle("Change in discharge rate")
ax.set_xlabel("Current")
ax.set_ylim(-0.0012, 0)
ax.set_ylabel("Slope")
ax.legend()
ax.grid("True")


num_predictions = 20
encoder_current_values = [0] * num_predictions
encoder_voltage_values = [3.8] * num_predictions
decoder_current_values = np.linspace(0.25, 3.5, num_predictions)
encoder_current, decoder_current, encoder_voltage, predicted_voltage = (
    perform_multiple_predictions(
        model_2,
        encoder_current_values,
        encoder_voltage_values,
        decoder_current_values,
        decoder_input_length=300,
    )
)

plot_prediction(
    encoder_current,
    decoder_current,
    encoder_voltage,
    predicted_voltage,
    title=f"Discharge Behavior at {num_predictions} different SOC/voltage levels",
)
_decoder_voltage = np.array([element[0] for element in predicted_voltage])

iR_drop = encoder_voltage_values - _decoder_voltage
ax.scatter(
    decoder_current_values,
    iR_drop,
    label=f"Encoder Voltage = {encoder_voltage_values[0]:.2f}",
)
ax.fill_betweenx([-0.05, 0.2], 2.5, 0.75, color="lightgray", alpha=0.5)
fig.suptitle("Pure Ohmic Drop")
ax.set_xlabel("Current")
ax.set_ylabel("Pure Ohmic Drop")
ax.legend()
ax.grid("True")



start = int(predicted_voltage[0].shape[0] / 3)

slopes = []
for i in range(num_predictions):
    slopes.append(np.polyfit(np.arange(start), predicted_voltage[i][-start:], 1)[0])
ax.scatter(
    decoder_current_values,
    slopes,
    label=f"Encoder Voltage = {encoder_voltage_values[0]:.2f}",
)


ax.fill_betweenx([-0.0012, 0], 2.5, 0.75, color="lightgray", alpha=0.5)
fig.suptitle("Change in discharge rate")
ax.set_xlabel("Current")
ax.set_ylim(-0.0012, 0)
ax.set_ylabel("Slope")
ax.legend()
ax.grid("True")
```
